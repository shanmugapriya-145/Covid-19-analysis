```python
# rename column iso 3166-1 alpha 3 CODE to Iso_Code
df = df.rename(columns={'ISO 3166-1 alpha-3 CODE': 'Iso_Code'})
df
```

```python
# top 5 countries which have more number of covid cases (hint use df.nlargest)
highest_covid_cases=df.nlargest(5, 'Total Cases')
highest_covid_cases
```

```python
#plot bar graph for top 5 countries

plt.bar(highest_covid_cases['Country'], highest_covid_cases['Total Cases'])
plt.xlabel('Country')
plt.ylabel('Number of COVID-19 Cases')
plt.title('Top 5 Countries with the Highest COVID-19 Cases')
plt.show()
```

```python
# least 5 countries which have least number of covid cases
least_covid_cases = df.nsmallest(5, 'Total Cases')
least_covid_cases
```

```python
#draw bar graph of least 5 countries
plt.bar(least_covid_cases['Country'], least_covid_cases['Total Cases'])
plt.xlabel('Country')
plt.ylabel('Number of COVID-19 Cases')
plt.title('Least 5 Countries with the Least COVID-19 Cases')
plt.show()
```

```python
#top 5 countries which have more number of covid Deaths
highest_covid_deaths=df.nlargest(5, 'Total Deaths')
highest_covid_deaths
```

```python
#plot bar graph for top 5 countries which have more number of covid Deaths
plt.bar(highest_covid_deaths['Country'], highest_covid_deaths['Total Deaths'])
plt.xlabel('Country')
plt.ylabel('Number of COVID-19 Deaths')
plt.title('Top 5 Countries with the Highest COVID-19 Deaths')
plt.show()
```

```python
#5 countries which have least number of covid Death
least_covid_deaths=df.nsmallest(5, 'Total Deaths')
least_covid_deaths
```

```python
#Death percentage
#top 5 countries which have more number of covid  Death percentage 
top_percent = df.nlargest(5, 'Death percentage')
top_percent
```

```python
#plot bar graph for top 5 countries which have more number of covid  Death percentage 
plt.bar(top_percent['Country'], top_percent['Death percentage'])
plt.xlabel('Country')
plt.ylabel('Death percentage')
plt.title('Top 5 Countries with the Most COVID-19 Death percentage')
plt.show()
```

```python
#continent wise cases (hint use group by)
cont_cases=df.groupby('Continent')['Total Cases'].sum().reset_index()
cont_cases
```

```python
#draw bar plot for continent wise case
plt.figure(figsize=(10,15))
plt.bar(cont_cases['Continent'], cont_cases['Total Cases'])
plt.xlabel('Continent')
plt.ylabel('Number of Cases')
plt.title('Continent-wise COVID-19 Cases')
plt.show()
```

```python
#find relationship between Population and total cases.Comment  on outliers
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Population', y='Total Cases', data=df)
plt.xlabel('Population')
plt.ylabel('Number of Cases')
plt.title('Relationship between Population and COVID-19 Cases')
plt.show()
```
